```python
lost_t = open('D://malcam3//GitHub//UM_openGH//study//mypage//team_project2//lostark_text.txt', 'rt', encoding='UTF-8').read()
```

```python
from konlpy.tag import Okt
from collections import Counter

okt = Okt()
line = []

line = okt.pos(lost_t)

line_t = []

for word, tag in line:
    if tag in ['Noun','Adjective']:
        line_t.append(word)
print(line_t)
```

```python
line_c = Counter(line_t)
line_c
```

```python
word_passed = {}
stop_words = ["욕설"]  # 불용어 : 욕설, 비하발언
for tag, counts in line_c.most_common(80):
    if tag in stop_words:
        continue
    if (len(tag) > 1):
        word_passed[tag] = counts
word_passed
```

```python
from wordcloud import WordCloud

wc = WordCloud(font_path=font_path, background_color="ivory", width=800, height=600)
cloud = wc.generate_from_frequencies(word_passed)
plt.imshow(cloud)
plt.axis("off")
plt.show()
```

```python
# 백준 토마토 문제
# BFS(deque)

from collections import deque # stack / queue 사용
     # 상 하 좌 우
dx = [ 0, 0, -1, 1]
dy = [-1, 1,  0, 0]

m, n = map( int, input().split() ) # 공간 설정
matrix = [ list(map( int, input().split() )) for i in range(n) ] # 리스트에 세로의 수만큼 반복 설정
q = deque([]) # 큐 실행
res = 0       # 결과값 초기화

for hang in range(n): # 0, 1, 2, 3
    for yeol in range(m): # 0, 1, 2, 3, 4, 5
        if matrix[hang][yeol] == 1:
            q.append([hang, yeol])

def bfs():
    while q:
        x, y = q.popleft()
        for i in range(4):
            nx, ny = x+dx[i], y+dy[i]  # 범위 밖으로 나가게 하지 않기 위해서 범위 설정
            if 0 <= nx < n and 0 <= ny < m and matrix[nx][ny] == 0:
                matrix[nx][ny] = matrix[x][y] + 1 # 익히기 / 다음 토마토 = 현재 토마토 + 1
                q.append([nx, ny]) # 큐 
bfs()
```

```python
matrix
```

```python
keep_going = True
for hang in matrix:
    for yeol in hang:
        if keep_going == True:
            if yeol == 0: # bfs로 못 익혔으면,
                print(-1)
                keep_going = False # exit(0)
    res = max(res, max(hang))
print(res - 1)
```
