```python
data = {
    "key1":["K0", "K0", "K1", "K2"], # 영등포구, 마포구
    "key2":["K0", "K1", "K0", "K1"], # 당산동, 문래동, 합정동
    "A":["A0", "A1", "A2", "A3"],
    "B":["B0", "B1", "B2", "B3"]
}
left = pd.DataFrame(data)
left
```

```python
data = {
    "key1":["K0", "K1", "K1", "K2"],
    "key2":["K0", "K0", "K0", "K0"],
    "C":["C0", "C1", "C2", "C3"],
    "D":["D0", "D1", "D2", "D3"]
}
right = pd.DataFrame(data)
right
```

```python
# sql 키
pd.merge(left, right, on=["key1", "key2"]) # how='inner'
```

```python
pd.merge(left, right, on=["key1", "key2"], how="left")
```

```python
pd.merge(left, right, on=["key1", "key2"], how="right")
```

```python
pd.merge(left, right, on=["key1", "key2"], how="outer")
```

```python
# for문 사용 하지말라고 경고문이 뜸
# 소수점 3번째 자리에서 반올림
for i in range(len(df_titanic)):
    df_titanic["fare"][i] = round(df_titanic["fare"][i], 2)
df_titanic
```

```python
# 일괄 적용
# 소수점 2번째 자리에서 반올림
def my_round(n):
    return round(n, 1)
df_titanic["fare"] = df_titanic["fare"].apply(my_round)
df_titanic
```

```python
# 소수점 1번째 자리에서 반올림
df_titanic["fare"] = df_titanic["fare"].apply(lambda x: round(x, 0))
df_titanic
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris.keys()  # 분류 문제: 문제들(data), 답들(target)
```

```python
len(iris.data)  # 150 송이의 꽃 정보
```

```python
iris.data[:2]
```

```python
iris.feature_names
```

```python
iris.target  # 150송이(3종류)
```

```python
iris.target_names
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logReg = LogisticRegression(max_iter=10000)
```

```python
logReg.fit(iris.data, iris.target)
```

```python
logReg.predict([[5.2, 3.4, 1.3, 0.1]]) # [5.1, 3.5, 1.4, 0.2] setosa:0
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(iris.data, iris.target)
```

```python
rf.predict([[5.2, 3.4, 1.3, 0.1]])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
logReg = LogisticRegression(max_iter=10000)
logReg.fit(X_train, y_train)
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
logReg.score(X_test, y_test), rf.score(X_test, y_test)
```

```python
from collections import Counter
```

```python
counts = Counter(news_nouns)
counts
```

```python
word_passed = {}
stop_words = ["개새끼", "시팔", "세종시"]  # 불용어 : 욕설, 비하발언
for tag, counts in counts.most_common(80):
    if tag in stop_words:
        continue
    if (len(tag) > 1):
        word_passed[tag] = counts
word_passed
```

```python
plt.figure(figsize=(10, 12))
plt.title("세종시 - 네이버 검색 결과")
plt.xlabel("counts")
plt.ylabel("keywords")

sorted_keys = sorted(word_passed, key=word_passed.get, reverse=False)
sorted_values = sorted(word_passed.values())

plt.barh(range(len(word_passed)), sorted_values)
plt.yticks(range(len(word_passed)), sorted_keys)
plt.show()
```
