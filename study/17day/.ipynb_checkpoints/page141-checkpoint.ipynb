```python
# data scrapping vs data crawling (데이터 크롤링)
```

```python
url_hollys = "https://www.hollys.co.kr/store/korea/korStore2.do?pageNo=51&sido=&gugun=&store="
print(url_hollys)
```

```python
for i in range(1, 52):
    url_hollys = f"https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={i}&sido=&gugun=&store="
    print(url_hollys)
```

```python
shop = []
for i in range(1, 52):
    url_hollys = f"https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={i}&sido=&gugun=&store="
    print(url_hollys)
    
    soup = BeautifulSoup(requests.get(url_hollys).text, "html.parser")
    tbody = soup.find("tbody")
    trs = tbody.find_all("tr")
    
    for tr in trs:
        tds = tr.find_all("td")
        local = tds[0].string
        shopname = tds[1].string
        address = tds[3].string
        phonenum = tds[5].string
        shop.append([local, shopname, address, phonenum])
```

```python
shop_length = len(shop)
print(shop_length)
```

```python
shop
```

```python
# !pip install pandas
import pandas as pd
```

```python
df_shop = pd.DataFrame(shop)
df_shop
```

```python
# incoding의 문제로 엑셀에서는 깨지고 메모장에서는 잘 나오는 현상을 encoding으로 엑셀에서 보이게 한다
df_shop.to_csv("hollys_shop.csv", encoding="cp949")
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
wd = webdriver.Chrome()
wd.get("https://www.coffeebeankorea.com/store/store.asp")
wd.execute_script("storePop2(31)") # 삼성봉은사거리점
```

```python
soup = BeautifulSoup(wd.page_source, "html.parser")
store_name = soup.select("div.store_txt>h2")[0].string
store_name
```

```python
store_info = soup.select("div.store_txt>table>tbody>tr>td")
store_info
```

```python
address = store_info[2].get_text().strip()
address
```

```python
phone_num = store_info[3].string
phone_num
```

```python
table_list = []
table_list.append([store_name, address, phone_num])
table_list
```
